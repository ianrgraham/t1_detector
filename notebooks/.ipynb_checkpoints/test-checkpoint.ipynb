```python
def plot_bd_pcurve_prog(dir, func=psc.pc_b):
    plt.figure()
    g2 = glob.glob(dir+"phom/*.npz")
    sorter = [int(g.split('/')[-1].replace("_bd.npz","")) for g in g2]
    nsort, ng2 = zip(*sorted(zip(sorter, g2)))
    bd1 = ng2[0] 
    a1 = np.load(bd1)['bd']
    a1 = a1[a1[:,0] > -1]
    cmap = mpl.cm.viridis
    norm = mpl.colors.Normalize(vmin=1, vmax=len(ng2)-1)
    x, s1 = psc.sum_persistence_func(a1[:,0].astype(np.float64),a1[:,1].astype(np.float64),func, points=np.linspace(-.5,1.5,101), os=0)
    ls = s1
    for i in np.arange(4,len(ng2),4):
        bd2 = ng2[i]

        a2 = np.load(bd2)['bd']
        a2 = a2[a2[:,0] > -1]
    
        _, s2 = psc.sum_persistence_func(a2[:,0].astype(np.float64),a2[:,1].astype(np.float64),func, points=np.linspace(-.5,1.5,101), os=0)
        print(np.mean(np.abs(ls-s2)))
        plt.plot(x, s2-s1, color=cmap(norm(i)))
        ls = s2
    
```

```python
def plot_bd_pcurve(dir):
    g2 = glob.glob(dir+"phom/*.npz")
    sorter = [int(g.split('/')[-1].replace("_bd.npz","")) for g in g2]
    nsort, ng2 = zip(*sorted(zip(sorter, g2)))
    bd1 = ng2[0] 
    bd2 = ng2[-1]
    a1 = np.load(bd1)['bd']
    a1 = a1[a1[:,0] > -1]
    a2 = np.load(bd2)['bd']
    a2 = a2[a2[:,0] > -1]
    
    log = os.path.join(dir, "log.dat")
    df = pd.read_csv(log, delimiter='\t')
    
    l = len(df.index)
    plt.figure()
    df.iloc[int(9/10*l):].potential_energy.plot()
    
    #make_hists(a1.T, a2.T, samples=1, log=False, bins=np.linspace(-1,1.5,71))

    print(a1[:,0].shape)
    #make_hists(a1.T, a2.T, samples=1, log=False, bins=np.linspace(-1,1.5,71))
    _, s1 = psc.sum_persistence_func(a1[:,0].astype(np.float64),a1[:,1].astype(np.float64),psc.pc_b, points=np.linspace(-.5,1.5,101), os=0)
    x, s2 = psc.sum_persistence_func(a2[:,0].astype(np.float64),a2[:,1].astype(np.float64),psc.pc_b, points=np.linspace(-.5,1.5,101), os=0)
    plt.figure()
    plt.plot(x, s1)
    plt.plot(x, s2)
    plt.figure()
    plt.plot(x, s2-s1)
```

```python
def make_hists(bd_untrained, bd_trained, samples, dim=0, log=True, zoom=False, bins=100):
    """Makes the histograms to be plotted for this topology exploration.
    
    Returns nothing.
    """
    
    fig = plt.figure(figsize=(5, 15))
    
    #fig, ax = plt.subplots(1,1, )
    
    ax1 = fig.add_subplot(3,1,1)
    ax2 = fig.add_subplot(3,1,2)
    ax3 = fig.add_subplot(3,1,3)
    
    ax1.set_xlabel(r"$Birth [\alpha]$")
    ax1.set_ylabel(r"$Death [\alpha]$")
    ax1.set_title("Untrained")
    
    ax2.set_xlabel(r"$Birth [\alpha]$")
    ax2.set_ylabel(r"$Death [\alpha]$")
    ax2.set_title("Trained")
    
    ax3.set_xlabel(r"$Birth [\alpha]$")
    ax3.set_ylabel(r"$Death [\alpha]$")
    ax3.set_title("Untrained (Red), Trained (Blue)")
    
    if zoom:
        for ax in [ax1, ax2, ax3]:
            ax.set_xlim(-0.5,0.1)
            ax.set_ylim(0.0,0.4)
    
    weights1 = np.ones_like(bd_untrained[0])/samples
    
    weights2 = np.ones_like(bd_trained[0])/samples
    
    btot = np.concatenate([bd_untrained[0], bd_trained[0]])
    
    dtot = np.concatenate([bd_untrained[1], bd_trained[1]])
    
    tmp =  weights2*-1
    
    wtot = np.concatenate([weights1,tmp])
    
    (hist, xedges, yedges) = np.histogram2d(bd_untrained[0], bd_untrained[1], weights=weights1, bins=bins, normed=False)

#     (hist, xedges, yedges) = np.histogram2d(np.log10(basin_pers), np.log10(crack_pers), bins=nbins,
#                                            range=[[np.log10(xmin), np.log10(xmax)], [np.log10(ymin), np.log10(ymax)]])

    hist = ma.masked_equal(hist, 0.0)

    #cmap = plt.get_cmap('viridis')#plt.cm.Reds
    cmap = plt.cm.Reds
    if log:
        norm = mpl.colors.LogNorm(vmin=np.min(np.abs(hist)), vmax=np.max(np.abs(hist)))
    else:
        norm = cm.colors.Normalize(vmin=np.min(np.abs(hist)), vmax=np.max(np.abs(hist)))
    smap = cm.ScalarMappable(norm=norm, cmap=cmap)

    X, Y = np.meshgrid(xedges, yedges)
#     X, Y = np.meshgrid(10**xedges, 10**yedges)
    im = ax1.pcolormesh(X, Y, hist.T, cmap=cmap, norm=norm)
    if dim > 0:
        ax1.plot(np.linspace(-.5,1.5,100), np.linspace(-.5,1.5,100), 'k--')
    fig.colorbar(im, ax=ax1)
    #ax1.pcolormesh(X, Y, hist.T, cmap=cmap, norm=norm)
    
    #******************ax1.hist2d(bd_untrained[0], bd_untrained[1], bins=50, weights=weights1)
    
    (hist, xedges, yedges) = np.histogram2d(bd_trained[0], bd_trained[1], weights=weights2, bins=bins, normed=False)

#     (hist, xedges, yedges) = np.histogram2d(np.log10(basin_pers), np.log10(crack_pers), bins=nbins,
#                                            range=[[np.log10(xmin), np.log10(xmax)], [np.log10(ymin), np.log10(ymax)]])

    hist = ma.masked_equal(hist, 0.0)

    #cmap = plt.get_cmap('viridis')#plt.cm.Blues
    cmap = plt.cm.Reds
    if log:
        norm = mpl.colors.LogNorm(vmin=np.min(np.abs(hist)), vmax=np.max(np.abs(hist)))
    else:
        norm = cm.colors.Normalize(vmin=np.min(np.abs(hist)), vmax=np.max(np.abs(hist)))
    smap = cm.ScalarMappable(norm=norm, cmap=cmap)

    X, Y = np.meshgrid(xedges, yedges)
#     X, Y = np.meshgrid(10**xedges, 10**yedges)

    im = ax2.pcolormesh(X, Y, hist.T, cmap=cmap, norm=norm)
    if dim > 0:
        ax2.plot(np.linspace(-.5,1.5,100), np.linspace(-.5,1.5,100), 'k--')
    fig.colorbar(im, ax=ax2)
    #**************ax2.hist2d(bd_trained[0], bd_trained[1], bins=50, weights=weights2)
    (hist, xedges, yedges) = np.histogram2d(btot, dtot, weights=wtot, bins=bins)

#     (hist, xedges, yedges) = np.histogram2d(np.log10(basin_pers), np.log10(crack_pers), bins=nbins,
#                                            range=[[np.log10(xmin), np.log10(xmax)], [np.log10(ymin), np.log10(ymax)]])

    hist = ma.masked_equal(hist, 0.0)
    
    print(np.min(np.abs(hist)), np.max(np.abs(hist)))

    cmap = plt.cm.Reds
    if log:
        norm = mpl.colors.LogNorm(vmin=np.min(np.abs(hist)), vmax=np.max(np.abs(hist)))
    else:
        norm = cm.colors.Normalize(vmin=np.min(np.abs(hist)), vmax=np.max(np.abs(hist)))
    smap = cm.ScalarMappable(norm=norm, cmap=cmap)

    X, Y = np.meshgrid(xedges, yedges)
#     X, Y = np.meshgrid(10**xedges, 10**yedges)
    im = ax3.pcolormesh(X, Y, ma.masked_less(hist.T, 0.0), cmap=cmap, norm=norm)
    fig.colorbar(im, ax=ax3)
    
    cmap = plt.cm.Blues
    im = ax3.pcolormesh(X, Y, ma.masked_less(-hist.T, 0.0), cmap=cmap, norm=norm)
    #plt.colorbar()
    if dim > 0:
        ax3.plot(np.linspace(-.5,1.5,100), np.linspace(-.5,1.5,100), 'k--')
    
    #cax = fig.add_axes([0.95, 0.11, 0.03, 0.77])
    #cbar = mpl.colorbar.ColorbarBase(cax, cmap=cmap, norm=norm, orientation='vertical')

    #cax.tick_params(which='major', direction='out', length=3.0, width=1.0)

    #cbar.set_label(r"Average Count")
    fig.colorbar(im, ax=ax3)
    
    #ax3.hist2d(btot, dtot, wtot)
    
    return
```
